# Aprendizagem de máquina

## Métodos Preditivos - Aprendizagem supervisionada
- Classificação: define classes para os registros exemplo o risco de credito (risco alto, medio, baixo). Cada registro pertence a uma classe (categoria) e possui um conjunto de atributos previsores, o objetivo é descobrir o relacionamento entre os atributos previsores e o atributo meta (este atributo meta tem valor conhecido já que a aprendizagem é supervisionada)
- Regressão: previsão de valores numericos ex: quanto a empresa ganha de lucro baseado nos dados históricos


##### Na aprendizagem supervisionada é feita a extração das características (dados) e o algoritimo de aprendizagem é feito com um supervisor, ou seja, será indicado qual o preço e localização do item A para treinar o modelo e realizar predições




## Métodos Descritivos - Aprendizagem não supervisionada
- Associação: se compra A então compra B ex: organização de prateleiras de mercado


- Agrupamento: encontrar grupos ex: segmentação de clientes (oferecer produto a um perfil especifico)


- Detecção de desvios: outliers ex: fraude em cartão de credito


- Padrões sequenciais: exemplo livrarias enviarem ofertas de livros parecidos com os que voce ja comprou antes

- Sumarização: reduzir informações, também pode ser usada para identificação de perfis na segmentação de clientes


##### Na aprendizagem não supervisionada é feita a extração das características (dados) e o algoritimo de aprendizagem é feito sem um supervisor, ou seja, os dados são analisados automaticamente para treinar o modelo, sem ensino prévio. Esse tipo precisa de análise para determinar o significado dos padrões encontrados
