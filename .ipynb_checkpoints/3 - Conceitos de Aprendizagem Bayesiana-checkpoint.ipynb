O algoritmo de Naive Bayes consiste em encontrar uma probabilidade a posteriori (possuir a doença, dado que recebeu um resultado positivo), multiplicando a probabilidade a priori (possuir a doença) pela probabilidade de “receber um resultado positivo, dado que tem a doença”. Esse algoritmo considera as variaveis independentes



Gerar uma tabela de probabilidades

 

Cálculos de probabilidade:
* Para esse caso abaixo há 79% de chance do risco ser baixo




Outro cliente: A Correção Laplaciana é necessária pois a porcentagem de todos daria zero. Para isso, é necessário adicionar um registro onde o valor esta zero, por exemplo, a probabilidade de ser ruim e baixo passa a ser 1/15.
OBS: a biblioteca do python ja vem com a correção




Conceitos:

*Probabilidades a priori (conhecemos com antecedencia, apenas olhando para a base)*
* Temos um conjunto de bolinhas e queremos prever se a do meio sera vermelho ou azul, sendo que a Prob(vermelho) é 7/12 e a Prob(azul) é 5/12.
* Algoritmo de bayes tem um parametro chamado raio para estimar a prob dentro dele, nele a prob(vermelho) é 3/7 e a prob(azul) é 1/5

*Probabilidade final (posteriori)*
* Pf(vermelho) = 7/12 x 3/7 = 0.25
* Pf(azul) = 5/12 x 1/5 = 0.08
Portando a bolinha é classificada como vermelha

















### Base de risco de crédito

```python
import pandas as pd
```

```python
base_risco = pd.read_csv('bases/risco_credito.csv')
base_risco
```

```python
x_risco = base_risco.iloc[:,0:4].values  #pegando todas as colunas menos a de risco
x_risco
```

```python
y_risco = base_risco.iloc[:,4].values #somente a coluna de risco
y_risco
```

```python
from sklearn.preprocessing import LabelEncoder #para evitar erros assumidos que uma coluna é mais relevante que a outra, quando na verdade elas não tem pesos
label_encoder_historia = LabelEncoder()
label_encoder_divida = LabelEncoder()
label_encoder_garantia = LabelEncoder()
label_encoder_renda = LabelEncoder()
```

```python
x_risco[:,0] = label_encoder_historia.fit_transform(x_risco[:,0])
x_risco[:,1] = label_encoder_divida.fit_transform(x_risco[:,1])
x_risco[:,2] = label_encoder_garantia.fit_transform(x_risco[:,2])
x_risco[:,3] = label_encoder_renda.fit_transform(x_risco[:,3])
```

```python
x_risco #como a base é pequena, não usaremos o one hot encoder para não aumentar o numero de colunas
```

```python
#salvando as variaveis para nao precisar fazer o processamento de novo
```

```python
import pickle
with open('risco_credito.pkl','wb') as f:
    pickle.dump([x_risco, y_risco], f)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
naive_risco = GaussianNB() #recebendo o objeto da classe na distribuição normail(gaussiana)
naive_risco.fit(x_risco, y_risco) #treinamento do algoritmo para realizar o calculo de probabilidades
```

```python
#historia boa (0), divida alta (0), garantias nenhuma (1), renda >35 (2)
#historia ruim (2), divida alta (0), garantias adequada (0), renda <15 (0)

previsao = naive_risco.predict([[0,0,1,2],[2,0,0,0]])
```

```python
previsao #classificação final de credito
```

```python
naive_risco.classes_ #quais sao os tipos
```

```python
naive_risco.class_count_ #quantos sao os tipos
```

```python
naive_risco.class_prior_ #quais porcentagem dos tipos
```

### Base de crédito

```python
with open('bases/credit.pkl', 'rb') as f: #rb é a leitura do arquivo
    x_credit_treinamento, x_credit_teste, y_credit_treinamento, y_credit_teste = pickle.load(f)
```

```python
x_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
x_credit_teste.shape, y_credit_teste.shape
```

```python
naive_credit = GaussianNB()
y_credit_treinamento=y_credit_treinamento.astype('int')
naive_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = naive_credit.predict(x_credit_teste)
```

```python
previsoes
```

```python
y_credit_teste #comparando para ver se o algoritimo acima acertou a previsão
```

```python
#comparando para ver se o algoritimo acima acertou a previsão
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_credit_teste, previsoes) #precentual de acerto de 93 por cento
```

```python
confusion_matrix(y_credit_teste, previsoes) #matriz de acertos classe por classe 
#matriz
#clientes que pagam          428 acertos                               8 erros (pagam mas foram classificados como não pagam)
#clientes que não pagam      23 erros(não pagam mas foram classificado como pagam)        41 acertos 
```

```python
from yellowbrick.classifier import ConfusionMatrix
```

```python
#uma melhor visualização da matriz
cm = ConfusionMatrix(naive_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```

```python
print(classification_report(y_credit_teste, previsoes)) 
#support: base de teste 
#recall: temos 436 registros que pagam e o algoritimo consegue encontrar corretamente 98% dos que pagam, e quando encontra 
#ele tem uma precisão de acerto de 95%
```
