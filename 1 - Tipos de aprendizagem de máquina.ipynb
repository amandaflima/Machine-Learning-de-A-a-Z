# Aprendizagem de mquina

## Mtodos Preditivos - Aprendizagem supervisionada
- Classificao: define classes para os registros exemplo o risco de credito (risco alto, medio, baixo). Cada registro pertence a uma classe (categoria) e possui um conjunto de atributos previsores, o objetivo  descobrir o relacionamento entre os atributos previsores e o atributo meta (este atributo meta tem valor conhecido j que a aprendizagem  supervisionada)
- Regresso: previso de valores numericos ex: quanto a empresa ganha de lucro baseado nos dados histricos


##### Na aprendizagem supervisionada  feita a extrao das caractersticas (dados) e o algoritimo de aprendizagem  feito com um supervisor, ou seja, ser indicado qual o preo e localizao do item A para treinar o modelo e realizar predies




## Mtodos Descritivos - Aprendizagem no supervisionada
- Associao: se compra A ento compra B ex: organizao de prateleiras de mercado


- Agrupamento: encontrar grupos ex: segmentao de clientes (oferecer produto a um perfil especifico)


- Deteco de desvios: outliers ex: fraude em carto de credito


- Padres sequenciais: exemplo livrarias enviarem ofertas de livros parecidos com os que voce ja comprou antes

- Sumarizao: reduzir informaes, tambm pode ser usada para identificao de perfis na segmentao de clientes


##### Na aprendizagem no supervisionada  feita a extrao das caractersticas (dados) e o algoritimo de aprendizagem  feito sem um supervisor, ou seja, os dados so analisados automaticamente para treinar o modelo, sem ensino prvio. Esse tipo precisa de anlise para determinar o significado dos padres encontrados
